## Imports

```python
import numpy as np
from matplotlib import pyplot, cm
from mpl_toolkits.mplot3d import Axes3D

import time
from numba import njit
from pyccel.epyccel import epyccel
from pyccel.decorators import types
```

## Pure python function

```python
def solve_2d_nonlinearconv_pure(u, un, v, vn, nt, dt, dx, dy, c):

    ###Assign initial conditions
    ##set hat function I.C. : u(.5<=x<=1 && .5<=y<=1 ) is 2
    u[int(.5 / dy):int(1 / dy + 1), int(.5 / dx):int(1 / dx + 1)] = 2
    ##set hat function I.C. : v(.5<=x<=1 && .5<=y<=1 ) is 2
    v[int(.5 / dy):int(1 / dy + 1), int(.5 / dx):int(1 / dx + 1)] = 2
    row, col = u.shape
    
    #fill the update of u and v
    u[0,:]=1
    u[ny-1,:]=1
    u[:,0]=1
    u[:,nx-1]=1
    for n in range(nt):
        for i in range(1,nx-2):
            for j in range(1,ny-2):
                u[i,j]=u[i,j]-u[i,j]*(dt/dx)*(u[i,j]-u[i-1,j])-v[i,j]*(dt/dy)*(u[i,j]-u[i,j-1])
                v[i,j]=v[i,j]-u[i,j]*(dt/dx)*(v[i,j]-v[i-1,j])-v[i,j]*(dt/dy)*(v[i,j]-v[i,j-1])

    return 0
```

## Numba accelerator

```python
@njit(fastmath=True)
def solve_2d_nonlinearconv_numba(u, un, v, vn, nt, dt, dx, dy, c):

    ###Assign initial conditions
    ##set hat function I.C. : u(.5<=x<=1 && .5<=y<=1 ) is 2
    u[int(.5 / dy):int(1 / dy + 1), int(.5 / dx):int(1 / dx + 1)] = 2
    ##set hat function I.C. : v(.5<=x<=1 && .5<=y<=1 ) is 2
    v[int(.5 / dy):int(1 / dy + 1), int(.5 / dx):int(1 / dx + 1)] = 2
    row, col = u.shape
    
    #fill the update of u and v
    u[0,:]=1
    u[ny-1,:]=1
    u[:,0]=1
    u[:,nx-1]=1
    for n in range(nt):
        for i in range(1,nx-2):
            for j in range(1,ny-2):
                u[i,j]=u[i,j]-u[i,j]*(dt/dx)*(u[i,j]-u[i-1,j])-v[i,j]*(dt/dy)*(u[i,j]-u[i,j-1])
                v[i,j]=v[i,j]-u[i,j]*(dt/dx)*(v[i,j]-v[i-1,j])-v[i,j]*(dt/dy)*(v[i,j]-v[i,j-1])
        
    return 0
```

## Pyccel accelerator

```python
@types('float[:,:]', 'float[:,:]','float[:,:]','float[:,:]', 'int', 'float', 'float','float', 'float')
def solve_2d_nonlinearconv_pyccel(u, un, v, vn, nt, dt, dx, dy, c):

    ###Assign initial conditions
    ##set hat function I.C. : u(.5<=x<=1 && .5<=y<=1 ) is 2
    u[int(.5 / dy):int(1 / dy + 1), int(.5 / dx):int(1 / dx + 1)] = 2
    ##set hat function I.C. : v(.5<=x<=1 && .5<=y<=1 ) is 2
    v[int(.5 / dy):int(1 / dy + 1), int(.5 / dx):int(1 / dx + 1)] = 2
    row, col = u.shape
    
    #fill the update of u and v
    u[0,:]=1
    u[col-1,:]=1
    u[:,0]=2
    u[:,row-1]=1
    for n in range(nt):
        for i in range(1,row-2):
            for j in range(1,col-2):
                u[i,j]=u[i,j]-u[i,j]*(dt/dx)*(u[i,j]-u[i-1,j])-v[i,j]*(dt/dy)*(u[i,j]-u[i,j-1])
                v[i,j]=v[i,j]-u[i,j]*(dt/dx)*(v[i,j]-v[i-1,j])-v[i,j]*(dt/dy)*(v[i,j]-v[i,j-1])
        
    return 0
```

## Inputs

```python
###variable declarations
nx = 100
ny = 100
nt = 60
c = 1
dx = 2 / (nx - 1)
dy = 2 / (ny - 1)
sigma = .2
dt = sigma * dx

x = np.linspace(0, 2, nx)
y = np.linspace(0, 2, ny)

u = np.ones((ny, nx)) ##create a 1xn vector of 1's
v = np.ones((ny, nx))
un = np.ones((ny, nx))
vn = np.ones((ny, nx))     

solve_2d_nonlinearconv_pure(u, un, v, vn, nt, dt, dx, dy, c)

fig = pyplot.figure(figsize=(7, 5), dpi=100)
ax = fig.gca(projection='3d')                      
X, Y = np.meshgrid(x, y)                            
surf = ax.plot_surface(X, Y, u, cmap=cm.viridis)


fig = pyplot.figure(figsize=(7, 5), dpi=100)
ax = fig.gca(projection='3d')
surf2 = ax.plot_surface(X, Y, v, cmap=cm.viridis)
```
